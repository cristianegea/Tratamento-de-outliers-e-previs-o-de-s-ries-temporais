# 1. Preparação do notebook

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
```

#2. Leitura dos dados

```python
# Abertura da caixa de diálogo para selecionar o arquivo
uploaded = files.upload()

# Obtenção do nome do arquivo selecionado
filename = list(uploaded.keys())[0]

# Leitura do arquivo para um DataFrame
df = pd.read_excel(filename)

df.head()
```

# 3. Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, QuantileTransformer, PowerTransformer
```

```python
# Função para normalizar os dados
def normalize_series(df):
    """
    Aplica diferentes técnicas de normalização a todas as colunas de um dataframe.
    Retorna um novo dataframe com as colunas normalizadas.
    """
    df_norm = pd.DataFrame(index=df.index)

    # Aplicar os métodos de normalização
    scalers = {
        'Min-Max Scaler': MinMaxScaler(),
        'Standard Scaler': StandardScaler(),
        'Robust Scaler': RobustScaler(),
        'Quantile Transformer': QuantileTransformer(),
        'Power Transformer': PowerTransformer(method='yeo-johnson')
    }

    for method, scaler in scalers.items():
        df_norm[method] = scaler.fit_transform(df)

    # Aplicar transformação logarítmica
    df_norm['Log Transformed'] = np.log1p(df.replace({0: np.nan}))  # Evitar log(0), que é indefinido

    return df_norm

# Aplicar a função a todas as séries temporais
df_normalized = {col: normalize_series(df[[col]]) for col in df.columns}

# Concatenar os resultados em um único dataframe
df_final = pd.concat(df_normalized, axis=1)
df_final
```

```python
# Criar gráficos para cada série temporal com identificação da série original
for col in df.columns:
    df_plot = df_normalized[col].copy()
    df_plot.insert(0, 'Original', df[col])  # Adicionando a série original

    # Criando um layout de 2 linhas e 4 colunas para os gráficos
    fig, axes = plt.subplots(2, 4, figsize=(18, 10), sharex=True)

    # Lista de métodos e títulos
    methods = ['Original', 'Min-Max Scaler', 'Standard Scaler', 'Robust Scaler', 'Log Transformed',
               'Quantile Transformer', 'Power Transformer']
    titles = [f'{col} - Série Original', 'Normalização: Min-Max Scaler', 'Normalização: Standard Scaler',
              'Normalização: RobustScaler', 'Normalização: Log Transform', 'Normalização: Quantile Transformer',
              'Normalização: Power Transformer']
    linestyles = ['-', '--', '-.', ':', 'dotted', '-', '--']

    # Criando os gráficos corretamente acessando as colunas do dataframe
    for i, (method, title, linestyle) in enumerate(zip(methods, titles, linestyles)):
        row, col = divmod(i, 4)  # Determina a posição na grade (2x4)
        axes[row, col].plot(df_plot.index, df_plot[method], label=title, linestyle=linestyle)
        axes[row, col].set_title(title)
        axes[row, col].legend()
        axes[row, col].grid(True)

    # Remover o subplot vazio (posição extra na grade 2x4 se houver)
    if len(methods) < 8:
        fig.delaxes(axes[1, 3])

    # Ajuste do layout para evitar sobreposição
    plt.tight_layout()
    plt.show()
```

O RobustScaler foi o método que melhor lidou com os outliers sem alterar o formato da curva. Ele usa a mediana e o intervalo interquartil (IQR) para normalizar os dados, tornando-o resistente a valores extremos, ao contrário do StandardScaler, que é sensível a outliers. A série mantém a mesma estrutura original, apenas ajustando a escala, sem distorções visíveis.
